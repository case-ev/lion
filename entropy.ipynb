```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# def entropy_change(soc, amplitude, k):
#     return -amplitude * np.cos(2 * np.pi * soc ** 2) + amplitude / (1 + np.exp(-k * (soc - 0.5)))


def entropy_change(soc, amplitude, kappa, mu, sigma, l=1):
    return amplitude * (np.exp(-(soc - mu)**2 / (2 * sigma)) / np.sqrt(2 * np.pi * sigma ** 2) - l * np.exp(-kappa * soc)) + 0.05e-3
```

```python
soc = np.linspace(0, 1, 100)
_AMPLITUDE = 0.04e-3
_KAPPA = 3
_MU = 0.4
_SIGMA = 0.05
_LAMBDA = 7

fig, ax = plt.subplots()

ax.plot(soc, 1e3 * entropy_change(soc, _AMPLITUDE, _KAPPA, _MU, _SIGMA, _LAMBDA), label=f"$A = {_AMPLITUDE}, \\kappa = {_KAPPA}, \\mu = {_MU}, \\sigma = {_SIGMA}, \\lambda = {_LAMBDA}$")

ax.set_ylabel("$\\partial_T V_{{oc}}$ (mV K$^{{-1}})$")
ax.set_xlabel("SoC (-)")
# ax.legend()

ax.grid(alpha=0.2)
```
